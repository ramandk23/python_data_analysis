```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/student_scores.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Drop unanmed column
```

```python
df = df.drop("Unnamed: 0", axis = 1)
df.head()
```

```python
# Change weekly study hours column
```

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct","5-10")
df.head()
```

```python
plt.figure(figsize= (5,5) )
plt.title("Gender Distrubution")
ax = sns.countplot(data = df, x="Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
# From the above chart we have analysed that the number of female students is more than the numbe rof male students
```

```python
plt.title("Relation between Parent's Education & Student's score")
gb_parent_edu = df.groupby("ParentEduc").agg({"MathScore":"mean" , "ReadingScore":"mean", "WritingScore":"mean"})
#print(gb_parent_edu)
sns.heatmap(gb_parent_edu , annot = True)
plt.show()
```

```python
# From the above chart we concluded that the parent's education have an impact on student's score
```

```python
plt.figure(figsize = (5,5))
plt.title("Reltaion b/w Parent'Marital Status on Student's score")
gb_parent_marital = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean" , "ReadingScore":"mean", "WritingScore":"mean"})
sns.heatmap(gb_parent_marital , annot= True)
plt.show()
```

```python
# From the above chart ,we conclude that there is negligible impact of parent's marital status on childre's score
```

```python
plt.title("Relation b/w student's sports practise and their score")
gb_student_practise = df.groupby("PracticeSport").agg({"MathScore":"mean", "ReadingScore":"mean", "WritingScore":"mean"})
sns.heatmap(gb_student_practise, annot = True)
plt.show()
```

```python
# From above chart we conclude that student's sports practise have some impact on their scores
```

```python
plt.title("Relation between student's study hours & their scores")
gb_study_hours = df.groupby("WklyStudyHours").agg({"MathScore":"mean", "ReadingScore":"mean", "WritingScore":"mean"})
sns.heatmap(gb_study_hours , cmap='mako', annot = True)
plt.show()
```

```python
# From the above chart we conclude that student who study more hours weekly have slight impact  on their scores
```

```python
sns.boxplot(data = df , x= "MathScore")
plt.show()
```

```python
# Distribution of Ethnic Group
```

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

label_list = ["Group A","Group B","Group C","Group D","Group E"]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.title("Distribution of Ethnic Groups")
plt.pie(mlist, labels = label_list, autopct = "%1.2f%%")
plt.show()

```
